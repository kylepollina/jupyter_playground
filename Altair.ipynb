```python
import pandas
import altair as alt
from vega_datasets import data
```

```python
iris = data.iris()
```

```python
alt.Chart(iris).mark_bar().encode(
    x='species',
    y='mean(sepalLength)'
).properties(
    width=400
)
```

```python
cars = data.cars()
```

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin'
).properties(
    width=800
)
```

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    column='Origin',
    color='Origin'
)
```

```python
selector = alt.selection_interval(encodings=['x'])

plot = alt.Chart(cars).mark_point().encode(
    x='Miles_per_Gallon',
    y='Horsepower',
    color=alt.condition(selector, 'Origin:N', alt.value('lightgrey'))
).add_selection(
    selector
)

bar = alt.Chart(cars).mark_bar().encode(
    x='count()',
    y='Origin:N',
    color='Origin:N'
).transform_filter(
    selector
)

plot & bar
```

```python
alt.Chart(cars).mark_point().encode(
    x = 'Displacement',
    y = 'Horsepower',
    color = 'Origin:N',
    column = 'Origin'
)
```

```python
flights = data.flights_5k()
```

```python
alt.Chart(flights).mark_bar().encode(
    x = 'destination',
    y = 'count()',
    color = 'destination:N'
)
```

```python
alt.Chart(flights).mark_bar().encode(
    x = alt.X('origin:N', sort='-y'),
    y = 'count()',
    color = 'origin:N'
)
```

```python
alt.Chart(flights).mark_bar().encode(
    x = alt.X('origin:N', sort = '-y'),
    y = 'mean(distance)',
    color = 'origin:N',
)
```

```python
alt.Chart(flights).mark_bar().encode(
    x = alt.X('destination:N', sort = '-y'),
    y = 'mean(distance)',
    color = 'destination:N',
)
```
